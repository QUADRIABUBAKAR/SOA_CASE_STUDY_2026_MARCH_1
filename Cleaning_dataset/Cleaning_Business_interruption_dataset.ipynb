## Joining Tables with `pd.merge()`

The `pd.merge()` function in pandas allows you to combine two DataFrames based on common columns or indices. It's similar to SQL JOIN operations.

Here are the key parameters:

-   `left`, `right`: The two DataFrames you want to merge.
-   `on`: The column name(s) to join on. If the column names are different in the left and right DataFrames, you can use `left_on` and `right_on`.
-   `how`: The type of merge to perform. Common options include:
    -   `'inner'` (default): Returns only the rows where the join key(s) are present in *both* DataFrames.
    -   `'outer'`: Returns all rows from both DataFrames, filling `NaN` for values that don't have a match.
    -   `'left'`: Returns all rows from the *left* DataFrame, with matching rows from the right. `NaN` for unmatched rows in the right.
    -   `'right'`: Returns all rows from the *right* DataFrame, with matching rows from the left. `NaN` for unmatched rows in the left.
-   `suffixes`: A tuple of string suffixes to apply to overlapping column names (other than the join key(s)) from the left and right DataFrames, respectively. This helps distinguish columns that exist in both tables but are not the join key.

```python
# Let's join the 'freq' and 'sev' tables.
# Both tables have 'policy_id', 'station_id', and 'solar_system' as common columns.
# Let's perform an inner join on 'policy_id' to combine information only for policies present in both datasets.
# We'll use suffixes to differentiate columns that appear in both dataframes but are not the merge key.
merged_df = pd.merge(freq, sev,
                   on='policy_id',
                   how='inner',
                   suffixes=('_freq', '_sev'))

print("Shape of the merged DataFrame:", merged_df.shape)
print("First 5 rows of the merged DataFrame:")
display(merged_df.head())
```
